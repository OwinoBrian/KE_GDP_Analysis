2. Trend Analysis

```python
import matplotlib.pyplot as plt
```

```python
# Plot the GDP trends for each industry
plt.figure(figsize=(12, 8))
for column in Current_Prices_df.columns[1:17]:  # Columns 1 to 17 correspond to industries
    plt.plot(Current_Prices_df['Filled_Year'] + Current_Prices_df['Quarters'] * 0.25, Current_Prices_df[column], marker='o', label=column)

plt.title('GDP Trends by Industry')
plt.xlabel('Year')
plt.ylabel('GDP at market prices')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=5)
plt.grid(True)
plt.show()
```

```python
# Convert the 'Filled_Year' column to integer
Current_Prices_df['Filled_Year'] = Current_Prices_df['Filled_Year'].astype(int)

# Calculate a new column for year progression including quarters
Current_Prices_df['YearProgression'] = Current_Prices_df['Filled_Year'] + Current_Prices_df['Quarters'] * 0.25

# Group the data by 'YearProgression' and sum the GDP values for each year and quarter
yearly_sum = Current_Prices_df.groupby('YearProgression').sum()

# Convert GDP values to millions for y-axis
yearly_sum['GDP at market prices'] = yearly_sum['GDP at market prices'] / 1000000

# Plot the yearly sum of GDPs
plt.figure(figsize=(10, 6))
plt.plot(yearly_sum.index, yearly_sum['GDP at market prices'], marker='o')
plt.title('Gross Domestic Product | Yearly Performance')
plt.xlabel('Year and Quarter')
plt.ylabel('GDP at market prices (Million KES)')
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.grid(True)
plt.show()
```

```python
data_2022 = Current_Prices_df[Current_Prices_df['Filled_Year'] == 2022]

# Plot the bar graph for GDP contribution per industry in 2022
plt.figure(figsize=(20, 10))
industries = data_2022.columns[1:17]  # Assuming columns 1 to 10 correspond to industries


plt.plot(industries, data_2022[industries].sum())
plt.title('GDP Contribution per Industry in 2022')
plt.xlabel('Industry')
plt.ylabel('GDP at market prices')
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.grid(True)
plt.show()
```

```python
# Calculate the quarterly growth rates for each sector
# Creating a new DataFrame to store the growth rates
growth_rates_df = pd.DataFrame()

# Iterate through each industry and calculate growth rates
industries = Current_Prices_df.columns[1:17]  # Assuming columns 1 to 10 correspond to industries

for industry in industries:
    # Calculate the percentage change in GDP from one quarter to the next
    growth_rates = Current_Prices_df[industry].pct_change() * 100
    
    # Store the growth rates in the new DataFrame
    growth_rates_df[industry] = growth_rates

# Drop the first row since it will have NaN values due to the percentage change calculation
growth_rates_df = growth_rates_df.dropna()

# Print the calculated growth rates
growth_rates_df
```
