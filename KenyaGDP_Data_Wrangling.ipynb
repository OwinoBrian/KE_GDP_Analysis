The data set has been retrieved from the Kenya National Bureau of Statistics' official website. https://www.knbs.or.ke/data-releases/ ; the specific document being the the recent Quarter 1, 2023 GDP report. The analysis will delve into the reasons why this particular document/source was chosen.


```python
import pandas as pd
import openpyxl
```

The source data is in PDF format. The conversion to excel was done using https://www.ilovepdf.com/pdf_to_excel . After conversion, we need to load the tables into data frames

```python
# Load data from Table 4 worksheet
df_table4 = pd.read_excel('First-Quarter-2023-GDP-Report (1).xlsx', sheet_name='Table 4')

# Load data from Table 6 worksheet
df_table5 = pd.read_excel('First-Quarter-2023-GDP-Report (1).xlsx', sheet_name='Table 5')
```

From the outputs, its evident that the rows 1-4 contains the yearly values while the the rest of the rows are the quarterly values. For the purpose of this analysis, the records will be combined into separated into quarterly and yearly averages.

```python
# Set the first row as column headers
df_table4.columns = df_table4.iloc[0]
df_table5.columns = df_table5.iloc[0]

# Drop the duplicated row (used as headers) from both DataFrames
df_table4 = df_table4[1:]
df_table5 = df_table5[1:]
```

```python
# Remove leading and trailing spaces from column names in both DataFrames
df_table4.columns = df_table4.columns.str.strip()
df_table5.columns = df_table5.columns.str.strip()
```

```python
df_table5
```

```python
# Combine the two DataFrames
combined_df = pd.concat([df_table4, df_table5], axis=1)

# Display the combined DataFrame
combined_df
```

```python
# Separate the yearly and quarterly data
yearly_data = combined_df.head(5)


# Display the separate DataFrames
print("Yearly Data:")
print(yearly_data)
```

4. Preparing Quarterly Data

```python
# Extract the remaining rows (excluding the first 5 rows)
quarterly_data = combined_df.iloc[5:]

# Display the first few rows of the quarterly_data DataFrame
print(quarterly_data.head())
```

```python
# Drop rows where all values are NaN
quarterly_data_2 = quarterly_data.dropna(how='all')

# Display the updated DataFrame
print(quarterly_data_2)
```

```python
import pandas as pd

# Assuming you already have the quarterly_data_2 DataFrame
# Create a new column "Filled_Year" to store the calculated years
quarterly_data_2["Filled_Year"] = None  # Initialize the new column

quarters_per_year = 4
current_year = 2018
quarter_counter = 1

for index, row in quarterly_data_2.iterrows():
    quarterly_data_2.at[index, "Filled_Year"] = current_year
    quarterly_data_2.at[index, "Quarters"] = quarter_counter
    
    quarter_counter += 1
    if quarter_counter > quarters_per_year:
        quarter_counter = 1
        current_year += 1

# Display the updated DataFrame
quarterly_data_2
```

```python
#Dropping 'year' column

quarterly_data_2.drop(columns="Year", inplace=True)

quarterly_data_2
```

```python
#Define File Path
csv_file_path = "KenyaGDP_quarterly_values_2018-2023.csv"


# Save the extracted DataFrame to a CSV file
quarterly_data_2.to_csv(csv_file_path, index=False)
```
